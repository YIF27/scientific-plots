# Activation energy of direct exchange and ring exchange in the xy plane

```python
import matplotlib.pyplot as plt

# Sample data for two stages
stages_2 = ['Stage 1', 'Stage 2']
values_stage1_2 = 0
values_stage2_2 = 5.07

# Sample data for five stages
stages_5 = ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 4', 'Stage 5']
values_5 = [0, 0.65, 0.75 + 0.65, 0.65 + 0.75 + 1.01, 0.65 + 0.75 + 1.01 + 0.65]

# Create a new figure
plt.figure(figsize=(6, 4))  # Adjust figure size as needed

# Plot for the two-stage scenario
# plt.title('Two Stages')
plt.xlabel('Stages')
plt.ylabel('Values')
plt.ylim(-1, 5.5)  # Adjust y-axis limits based on your data ranges
plt.xticks([0, 1], stages_2)
plt.hlines(values_stage1_2, 0, 0.5, colors='orange', label='Direct Diffusion')
plt.hlines(values_stage2_2, 0.5, 1, colors='orange')
plt.plot([0.5, 0.5], [values_stage1_2, values_stage2_2], color='orange', linestyle='--')

# Add text annotations for each stage's value

plt.text(0.55, 4, 5.07, color='orange', ha='center')

plt.legend(loc='upper left')

# Plot for the five-stage scenario
# plt.title('Five Stages')
plt.xlabel('Stages')
plt.ylabel(r"Activation Energy (eV)")
plt.ylim(-0.5,5.5)  # Adjust y-axis limits based on your data ranges
plt.xticks([(i + 0.5) * 0.2 for i in range(len(stages_5))], stages_5)
for i, value in enumerate(values_5):
    plt.hlines(value, i * 0.2, (i + 1) * 0.2, colors='blue')
plt.hlines(0, 0, 0.2, colors='blue', label="Vacancy Diffusion")
for i in range(len(values_5) - 1):
    plt.plot([(i+1) * 0.2 , (i + 1) * 0.2 ], [values_5[i], values_5[i + 1]], color='blue', linestyle='--')
# plt.plot([1, 1], [values_stage1_2, values_stage2_2], color='green', linestyle='--')
plt.plot([1, 1], [0, 3.06], color='blue', linestyle='--')
plt.legend(loc='upper left')

# Add text annotations for each stage's value
for i, value in enumerate([0.65, 0.75, 1.01, 0.65]):
  x_axis = [0.1, 0.3, 0.5, 0.7]
  y_axis = [0, 0.65, 0.75 + 0.65, 0.65 + 0.75 + 1.01, 0.65 + 0.75 + 1.01 + 0.65]
  plt.text(x_axis[i], y_axis[i]+0.05, str(value), color='blue', ha='center')
plt.text(0.96, 2, str(3.06), color='blue', ha='center')
# Display the plot
# plt.tight_layout()
# Save the figure
plt.title("Activation energy of diffusion in xy plane")
plt.savefig('stages_plot.png', dpi =600)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
stages = ['Stage1', 'Stage2', 'Stage3', 'Stage4', 'Stage5']
activation_energies = [0.35, 0.42, 0.59, 0.35]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8,6))

# Plot the stages and activation energies
x0 = np.linspace(0, 4, 1000)
y0 = 2.66 * np.sin(np.pi * (x0/4))
ax.plot(x0, y0, 'k-', linewidth=2, label = "Direct Exchange")
ax.text(1.9, 2.46, "2.66", fontsize = 12)


for i, (stage, energy) in enumerate(zip(stages, activation_energies)):
      x = np.linspace(i, i+1, 1000)
      y = energy * np.sin(np.pi * (x-i))
      if i == 0:
        ax.plot(x, y, 'r-', linewidth=2, label = "Ring Diffusion")
        ax.text(i+0.38, energy-0.1, str(energy), fontsize = 12)
      else:
        ax.plot(x, y, 'r-', linewidth=2)
        ax.text(i+0.38, energy-0.1, str(energy), fontsize = 12)


# Set xticks and yticks
ax.set_xticks(range(len(stages)))
ax.set_xticklabels(stages)
ax.set_yticks([0, 0.5, 1.0, 1.5, 2.0, 2.5])
ax.set_yticklabels(['0', '0.5', '1.0', '1.5', '2.0', '2.5'])

# Set the title and y-label
ax.set_title("(001) Surface")
ax.set_ylabel("Activation Energy (eV)")
ax.legend()
# Show the plot
plt.tight_layout()
# plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.savefig('001-surface', dpi =600)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
stages = ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 4', 'Stage 5', 'Stage 6', 'Stage 7']
activation_energies = [1.43, 0.97, 1.16, 0.64, 0.71, 0.63]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8,6))

# Plot the stages and activation energies of direct exhcnage
x0 = np.linspace(0, 6, 1000)
y0 = 6.5 * np.sin(np.pi * (x0/6))
ax.plot(x0, y0, 'k-', linewidth=2, label = "Vacancy Diffusion")
ax.text(3, 6.2, "6.5", fontsize = 12)

# Plot the stages and activation energies of vacancy diffusion
x0 = np.linspace(0, 6, 1000)
y0 = 2.65 * np.sin(np.pi * (x0/6))
ax.plot(x0, y0, 'b-', linewidth=2, label = "Direct Exchange")
ax.text(3, 2.40, "2.65", fontsize = 12)


for i, (stage, energy) in enumerate(zip(stages, activation_energies)):
      x = np.linspace(i, i+1, 1000)
      y = energy * np.sin(np.pi * (x-i))
      if i == 0:
        ax.plot(x, y, 'r-', linewidth=2, label = "Ring Diffusion")
        ax.text(i+0.38, energy-0.3, str(energy), fontsize = 12)
      else:
        ax.plot(x, y, 'r-', linewidth=2)
        ax.text(i+0.35, energy-0.45, str(energy), fontsize = 12)


# Set xticks and yticks
ax.set_xticks(range(len(stages)))
ax.set_xticklabels(stages)
# ax.set_yticks([0, 0.5, 1.0, 1.5, 2.0, 2.5])
# ax.set_yticklabels(['0', '0.5', '1.0', '1.5', '2.0', '2.5'])

# Set the title and y-label
ax.set_title("(100) Surface")
ax.set_ylabel("Activation Energy (eV)")
ax.legend()
# Show the plot
plt.tight_layout()
# plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.savefig('100-surface.png', dpi =600)
plt.show()
```
