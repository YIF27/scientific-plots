# Activation energy of direct exchange and ring exchange in the xy plane

```python
import matplotlib.pyplot as plt

# Sample data for two stages
stages_2 = ['Stage 1', 'Stage 2']
values_stage1_2 = 0
values_stage2_2 = 5.07

# Sample data for five stages
stages_5 = ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 4', 'Stage 5']
values_5 = [0, 0.65, 0.75 + 0.65, 0.65 + 0.75 + 1.01, 0.65 + 0.75 + 1.01 + 0.65]

# Create a new figure
plt.figure(figsize=(6, 4))  # Adjust figure size as needed

# Plot for the two-stage scenario
# plt.title('Two Stages')
plt.xlabel('Stages')
plt.ylabel('Values')
plt.ylim(-1, 5.5)  # Adjust y-axis limits based on your data ranges
plt.xticks([0, 1], stages_2)
plt.hlines(values_stage1_2, 0, 0.5, colors='orange', label='Direct Diffusion')
plt.hlines(values_stage2_2, 0.5, 1, colors='orange')
plt.plot([0.5, 0.5], [values_stage1_2, values_stage2_2], color='orange', linestyle='--')

# Add text annotations for each stage's value

plt.text(0.55, 4, 5.07, color='orange', ha='center')

plt.legend(loc='upper left')

# Plot for the five-stage scenario
# plt.title('Five Stages')
plt.xlabel('Stages')
plt.ylabel(r"Activation Energy (eV/$\AA$^2 )")
plt.ylim(-0.5,5.5)  # Adjust y-axis limits based on your data ranges
plt.xticks([(i + 0.5) * 0.2 for i in range(len(stages_5))], stages_5)
for i, value in enumerate(values_5):
    plt.hlines(value, i * 0.2, (i + 1) * 0.2, colors='blue')
plt.hlines(0, 0, 0.2, colors='blue', label="Vacancy Diffusion")
for i in range(len(values_5) - 1):
    plt.plot([(i+1) * 0.2 , (i + 1) * 0.2 ], [values_5[i], values_5[i + 1]], color='blue', linestyle='--')
# plt.plot([1, 1], [values_stage1_2, values_stage2_2], color='green', linestyle='--')
plt.plot([1, 1], [0, 3.06], color='blue', linestyle='--')
plt.legend(loc='upper left')

# Add text annotations for each stage's value
for i, value in enumerate([0.65, 0.75, 1.01, 0.65]):
  x_axis = [0.1, 0.3, 0.5, 0.7]
  y_axis = [0, 0.65, 0.75 + 0.65, 0.65 + 0.75 + 1.01, 0.65 + 0.75 + 1.01 + 0.65]
  plt.text(x_axis[i], y_axis[i]+0.05, str(value), color='blue', ha='center')
plt.text(0.96, 2, str(3.06), color='blue', ha='center')
# Display the plot
# plt.tight_layout()
# Save the figure
plt.title("Activation energy of diffusion in xy plane")
plt.savefig('stages_plot.png', dpi =600)
plt.show()
```

# Activation energy of direct exchange and ring exchange in the xz plane

```python
import matplotlib.pyplot as plt

# Sample data for two stages
stages_2 = ['Stage 1', 'Stage 2']
values_stage1_2 = 0
values_stage2_2 = 6.5

# Sample data for two stages
stages_3 = ['Stage 1', 'Stage 2']
values_stage1_3 = 0
values_stage2_3 = 2.65



# Sample data for six stages
stages_7 = ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 4', 'Stage 5', 'Stage 6', 'Stage 7']
values_7 = [0, 1.43, 1.43 + 0.97, 1.43 + 0.97 + 1.16, 1.43 + 0.97 + 1.16+0.64, 1.43 + 0.97 + 1.16+0.64+0.71, 1.43 + 0.97 + 1.16+0.64+0.71+0.63]

# Create a new figure
plt.figure(figsize=(6, 4))  # Adjust figure size as needed

# Plot for the two-stage scenario
# plt.title('Two Stages')
plt.xlabel('Stages')
plt.ylabel('Values')
plt.ylim(-1, 6.5)  # Adjust y-axis limits based on your data ranges
plt.xticks([0, 1.4], stages_2)
plt.hlines(values_stage1_2, 0, 0.7, colors='orange', label='Direct Diffusion')
plt.hlines(values_stage2_2, 0.7, 1.4, colors='orange')
plt.plot([0.7, 0.7], [values_stage1_2, values_stage2_2], color='orange', linestyle='--')

# Plot for the two-stage scenario
# plt.title('Two Stages')
plt.xlabel('Stages')
plt.ylabel('Values')
plt.ylim(-1, 6.5)  # Adjust y-axis limits based on your data ranges
plt.xticks([0, 1.4], stages_2)
plt.hlines(values_stage1_3, 0, 0.7, colors='red', label='Vacancy Diffusion')
plt.hlines(values_stage2_3, 0.7, 1.4, colors='red')
plt.plot([0.7, 0.7], [values_stage1_3, values_stage2_3], color='red', linestyle='--')
plt.text(0.66, 1.35, str(2.65), color='red', ha='center')





# Add text annotations for each stage's value

plt.text(0.55, 4, 6.50, color='orange', ha='center')

plt.legend(loc='upper left')

# Plot for the five-stage scenario
# plt.title('Five Stages')
plt.xlabel('Stages')
plt.ylabel(r"Activation Energy (eV/$\AA$^2 )")
plt.ylim(-0.5,7)  # Adjust y-axis limits based on your data ranges
plt.xticks([(i + 0.5) * 0.2 for i in range(len(stages_7))], stages_7)
for i, value in enumerate(values_7):
    plt.hlines(value, i * 0.2, (i + 1) * 0.2, colors='blue')
plt.hlines(0, 0, 0.2, colors='blue', label="Ring diffusion with a vacancy")
for i in range(len(values_7) - 1):
    plt.plot([(i+1) * 0.2 , (i + 1) * 0.2 ], [values_7[i], values_7[i + 1]], color='blue', linestyle='--')
# plt.plot([1, 1], [values_stage1_2, values_stage2_2], color='green', linestyle='--')
plt.plot([1.4, 1.4], [0, 5.54], color='blue', linestyle='--')
plt.legend(loc='upper left')

# Add text annotations for each stage's value
for i, value in enumerate([1.43, 0.97, 1.16, 0.64, 0.71, 0.63]):
  x_axis = [0.1, 0.3, 0.5, 0.7, 0.9, 1.1]
  y_axis = [0, 1.43, 1.43 + 0.97, 1.43 + 0.97 + 1.16, 1.43 + 0.97 + 1.16+0.64, 1.43 + 0.97 + 1.16+0.64+0.71, 1.43 + 0.97 + 1.16+0.64+0.71+0.63]
  plt.text(x_axis[i], y_axis[i]+0.2, str(value), color='blue', ha='center')
plt.text(1.36, 2, str(5.53), color='blue', ha='center')
# Display the plot
# plt.tight_layout()
# Save the figure
plt.title("Activation energy of diffusion in xz plane")
plt.savefig('stages_plot.png', dpi =600)
plt.show()
```
