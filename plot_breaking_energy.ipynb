# Breaking energy after removing Zn

```python
import matplotlib.pyplot as plt

a = [0, 0.25, 0.50, 0.75, 1] # Zn content
E_001 = [0.22, 0.18, 0.13, 0.09, 0.04] # breaking energy along direction [001]
E_100 = [0.20, 0.18, 0.16, 0.14, 0.11] # breaking energy along direction [100]
E_111 = [0.16, 0.15, 0.13, 0.11, 0.08] # breaking energy along direction [111]

# Line plots
plt.plot(a, E_001, label='E_001', linestyle='-', marker='o')
plt.plot(a, E_100, label='E_100', linestyle='-', marker='s')
plt.plot(a, E_111, label='E_111', linestyle='-', marker='^')

# Scatter plots
plt.scatter(a, E_001, label='', marker='o', color='blue')
plt.scatter(a, E_100, label='', marker='s', color='green')
plt.scatter(a, E_111, label='', marker='^', color='red')

plt.xlabel('Zn content')
plt.ylabel('Breaking Energy (eV/Å²)')
plt.title('Breaking energy after removing Zn')
plt.legend()
# plt.grid(True)
# Save the figure with 600 dpi
plt.savefig('energy_vs_a.png', dpi=600)
plt.show()
```
